```python
import matplotlib.pyplot as plt

x1 = list(range(1, 21))
x2 = [i**2 for i in x1]
x3 = [i**0.5 * 10 for i in x1]
x4 = [i*3 + 5 for i in x1]
plt.figure(figsize=(10,5))
plt.subplot(2,2,1)
plt.plot(x1,x2)
plt.title("First Plot")

plt.subplot(2,2,2)
plt.title("Second Plot")
plt.plot(x1,x3)

plt.subplot(2,2,3)
plt.plot(x4,x1)
plt.title("Third Plot")
plt.subplot(2,2,4)

plt.subplot(2,2,4)
plt.plot(x2,x4)
plt.title("Fourth Plot")
```
