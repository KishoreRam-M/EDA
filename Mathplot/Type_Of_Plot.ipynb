```python
import matplotlib.pyplot as plt
import pandas
## bar chart
x=["iphone","samsung","android","Redmi","nothing"]
y=[200,500,100,50,250]
plt.bar(x,y,color='grey',linewidth=1)
plt.title("Bar Chart")
plt.xlabel("Mobile")
plt.ylabel("Growth")
plt.show()
```

```python
import pandas as pd

data = {
    'Region': ['North', 'South', 'East', 'West'],
    'Sales_Q1': [150, 200, 300, 250],
    'Sales_Q2': [180, 210, 320, 270]
}
df = pd.DataFrame(data)



plt.figure(figsize=(6,4))
plt.bar(df['Region'], df['Sales_Q1'], color='skyblue')
plt.title('Q1 Sales by Region')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.xticks(rotation=30)  # handles long category names
plt.tight_layout()
plt.savefig('bar_chart.png', dpi=150, bbox_inches='tight')
plt.show()
```
