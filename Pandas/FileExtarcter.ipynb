```python
import zipfile
import pdfplumber
import pandas as pd
from io import BytesIO

zip_path = r"C:\Users\KRM\Downloads\Internship.zip"

pdf_data = []
error_log = []

with zipfile.ZipFile(zip_path, 'r') as z:
    pdf_files = [f for f in z.namelist() if f.lower().endswith('.pdf')]

    for pdf_file in pdf_files:
        try:
            with z.open(pdf_file) as zf:
                pdf_bytes = BytesIO(zf.read())

                with pdfplumber.open(pdf_bytes) as pdf:
                    text = "\n".join(page.extract_text() or "" for page in pdf.pages)

                    if text.strip():
                        pdf_data.append({'filename': pdf_file, 'text': text})
                    else:
                        warning_msg = f"No text found in {pdf_file}"
                        print("Warning:", warning_msg)
                        error_log.append({'filename': pdf_file, 'error': 'Empty PDF / No text'})

        except Exception as e:
            print(f"Error reading {pdf_file}: {e}")
            error_log.append({'filename': pdf_file, 'error': str(e)})

df_pdfs = pd.DataFrame(pdf_data)
df_errors = pd.DataFrame(error_log)

print("Successfully read PDFs:")
print(df_pdfs.head(), "\n")

print("Failed / empty PDFs:")
print(df_errors.head(), "\n")
```
