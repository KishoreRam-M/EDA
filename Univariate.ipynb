# 📊 Histogram vs Pie Chart vs Count Plot

These three plots look similar sometimes, but each serves a **different purpose**. All of them are **univariate analysis tools** (focus on one variable at a time).

---

## 🔹 1. Histogram

* **What it is**:
  A plot showing the **distribution of continuous numerical data**.
  It divides values into **bins (intervals)** and counts how many values fall into each bin.

* **When to use**:

  * Understand the **shape of data** (normal, skewed, uniform).
  * Detect **spread and outliers** in numeric data.

* **Example**: Distribution of students’ marks, people’s heights, salary ranges.

* **Unique quality**:

  * Works with **continuous numeric variables**.
  * Groups data into **intervals (bins)**, not categories.

✅ **Use case**: *“How many students scored between 40–50, 50–60, etc.?”*

---

## 🔹 2. Pie Chart

* **What it is**:
  A **circular chart** divided into slices that represent the **proportion (percentage)** of categories.

* **When to use**:

  * Show **part-to-whole relationships**.
  * Compare percentages when there are **few categories (2–6 is ideal)**.

* **Example**: Market share of companies, percentage of voters by political party.

* **Unique quality**:

  * Works with **categorical variables** (labels).
  * Focuses on **percentage contribution** of each category.

✅ **Use case**: *“What percentage of sales comes from Product A vs Product B?”*

---

## 🔹 3. Count Plot

* **What it is**:
  A **bar plot** that shows the **frequency (count)** of each category in a categorical variable.

* **When to use**:

  * Check the **distribution of categories**.
  * Identify **imbalances** (e.g., dataset has more males than females).

* **Example**: Number of passengers in each class (1st, 2nd, 3rd), number of employees in each department.

* **Unique quality**:

  * Works with **categorical/discrete variables**.
  * Shows **raw counts** (can be normalized to percentages if needed).

✅ **Use case**: *“How many males vs females are in the dataset?”*

---

## 🔑 Quick Comparison Table

| Chart Type     | Data Type                            | Shows                       | Best For                                  | Unique Quality                        |
| -------------- | ------------------------------------ | --------------------------- | ----------------------------------------- | ------------------------------------- |
| **Histogram**  | Continuous numerical                 | Distribution across ranges  | Spread, shape, outliers                   | Groups data into bins (intervals)     |
| **Pie Chart**  | Categorical + numerical (proportion) | Percent share of categories | Small number of categories, part-to-whole | Circular % visualization              |
| **Count Plot** | Categorical (discrete)               | Frequency of categories     | Class balance, comparing categories       | Bar chart showing raw category counts |

---

## 🎯 Easy Way to Remember

* **Histogram** → *“How is numeric data distributed?”*
* **Pie Chart** → *“What percentage does each category contribute?”*
* **Count Plot** → *“How many items in each category?”*
